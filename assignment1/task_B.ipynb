# Intro to ML : Exercise B

1). Dataset 1 creation. The dataset shall represent a binary classification problem and
shall have as input two continuous real number attributes. The dataset shall have 40 data points
belonging to one class and 40 data points belonging to the other class. Discuss your approach and
report the data points obtained in the report.

```python
from sklearn.datasets import make_classification
import numpy as np

# Dataset 1 creation
X1, y1 = make_classification(n_samples=80, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, random_state=1)

print("Data points for Dataset 1:")
for i in range(len(X1)):
    print(f"{X1[i]}, {y1[i]}")
```

2). Dataset 2 creation (Dataset 1 + outliers). Add randomly to Dataset 1, four outliers for
each class. Discuss your approach and report the data points, including the outliers in the report

```python
# Adding outliers to Dataset 1
np.random.seed(2)  # for reproducibility
outliers1 = np.random.uniform(low=-3, high=3, size=(4, 2))  # Generating outliers for class 0
outliers2 = np.random.uniform(low=-3, high=3, size=(4, 2))  # Generating outliers for class 1

X2 = np.concatenate((X1, outliers1, outliers2), axis=0)
y2 = np.concatenate((y1, np.zeros(4), np.ones(4)))  # Assuming 0 represents one class and 1 represents the other class

print("\nData points for Dataset 2 (including outliers):")
for i in range(len(X2)):
    print(f"{X2[i]}, {y2[i]}")
```

3). Split Dataset 1 and Dataset 2 in training and testing data. Justify your choices for
splitting the data and report the split data.

```python
from sklearn.model_selection import train_test_split

# Splitting Dataset 1 into training and testing data
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=1)

# Splitting Dataset 2 into training and testing data
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=1)

# Report the split data
print("Split data for Dataset 1:")
print(f"X1_train shape: {X1_train.shape}, y1_train shape: {y1_train.shape}")
print(f"X1_test shape: {X1_test.shape}, y1_test shape: {y1_test.shape}\n")

print("Split data for Dataset 2:")
print(f"X2_train shape: {X2_train.shape}, y2_train shape: {y2_train.shape}")
print(f"X2_test shape: {X2_test.shape}, y2_test shape: {y2_test.shape}\n")
```

4). Train on Dataset 1 and Dataset 2, the following classification models: k-NN, Naive Bayes,
Decision Trees, and Random Forests. Report the classification accuracy obtained on the training
and on the testing data. Report also the confusion matrix. Discuss the obtained results.

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Initializing classifiers
knn = KNeighborsClassifier()
nb = GaussianNB()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()

# Training models on Dataset 1
knn.fit(X1_train, y1_train)
nb.fit(X1_train, y1_train)
dt.fit(X1_train, y1_train)
rf.fit(X1_train, y1_train)

# Training models on Dataset 2
knn.fit(X2_train, y2_train)
nb.fit(X2_train, y2_train)
dt.fit(X2_train, y2_train)
rf.fit(X2_train, y2_train)

# Prediction on test data
y1_pred = knn.predict(X1_test)
y2_pred = knn.predict(X2_test)

# Calculating accuracy and confusion matrix
accuracy1 = accuracy_score(y1_test, y1_pred)
accuracy2 = accuracy_score(y2_test, y2_pred)

conf_matrix1 = confusion_matrix(y1_test, y1_pred)
conf_matrix2 = confusion_matrix(y2_test, y2_pred)


print("Results for Dataset 1:")
print(f"Accuracy: {accuracy1}")
print("Confusion Matrix:")
print(conf_matrix1)
print("\n")

print("Results for Dataset 2:")
print(f"Accuracy: {accuracy2}")
print("Confusion Matrix:")
print(conf_matrix2)
```

5). Taking inspiration from the tutorial mentioned above, visualise the decision boundaries
of the various classification models used.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Helper function to plot the decision boundaries
def plot_decision_boundaries(classifier, X, y, title):
    h = .02  # step size in the mesh

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Plot the decision boundaries
    fig, ax = plt.subplots()
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax.set_title(title)

    # Plot the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cm_bright, edgecolors='k')

    # Create a mesh to plot in
    mesh = np.c_[xx.ravel(), yy.ravel()]
    if hasattr(classifier, "decision_function"):
        Z = classifier.decision_function(mesh)
    else:
        Z = classifier.predict_proba(mesh)[:, 1]

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=cm, alpha=0.8)

    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())

# Visualizing decision boundaries for each classifier on Dataset 1
classifiers = [knn, nb, dt, rf]
titles = ['K-NN', 'Naive Bayes', 'Decision Trees', 'Random Forests']
datasets = [(X1_train, y1_train)]

for ds in datasets:
    X, y = ds
    for i, clf in enumerate(classifiers):
        plot_decision_boundaries(clf, X, y, titles[i])
        plt.show()
```
