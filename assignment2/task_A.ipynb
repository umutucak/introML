```python
# -*- coding: utf-8 -*-
# Revised version 02/11/2023

import numpy as np 
import time
import random
import matplotlib.pyplot as plt

# compute a reward based on a probability
def get_reward( prob, n = 10 ):
    reward = 0
    for i in range( n ):
        if random.random() < prob: # random() returns a float < 1.0
            reward += 1
    return reward

# find the best machine based on past rewards
def get_best_arm( pastRewards, actions ):
    bestArm = 0 # just default to arm #0 at the beginning
    bestAvg = 0
    avg = 0
    for action in actions:
        #TODO:
        #step1: find in the history the index of all experiments where "action" was performed
        tmp = np.where(pastRewards[:,0] == action)
        #step2: compute the mean reward over these experiments
        avg = np.mean(pastRewards[tmp,1])
        if avg > bestAvg:
            bestAvg = avg
            bestArm = action
    print("bestArm:", bestArm,  " avg:", bestAvg )
    return bestArm

nb_machines = 10
# set fixed seed as machines' probabilities do not change over experiments
np.random.seed(0)
# chance to win for each machine
#   DO NOT access this information to improve your strategy, that would be considered cheating
#   you can use the information to verify wether your algorithm learned the right choice or not though
arms = np.random.rand( nb_machines ) # create a collection of 10 random float between 0 and 1
print( arms )

# reset to the seed to have actual randomness for each pull of an arm
np.random.seed()

# Number of trials in a single experiment
nb_trials = 1000

starting_time=time.time()

# Numpy-Array of 2-dimensional vectors, initialized with zeros
# first dimension: action taken, second dimension: reward
# both dimensions contain integer values
pastRewards = np.zeros( ( nb_trials, 2 ) ) 

plt.xlabel( "Trials" )
plt.ylabel( "Avg Reward" )

best_arm=0

num_experiments = 10
all_past_rewards = []

for experiment in range(num_experiments):
    epsilon = 0.1

for i in range( nb_trials ):
    # TODO: use the "random.random()" and "get_best_arm(...)" methods to implement the epsilon-greedy strategy
    if random.random() < epsilon:
        choice = np.random.choice(nb_machines)  # Exploration: We randomly choose an arm
    else:
        best_arm = get_best_arm(pastRewards, range(nb_machines))
        choice = best_arm  # Exploitation: We choose the arm with best average reward

    
    # pull the arm of machine with id [choice] and collect the wins
    reward = get_reward( arms[ choice ] )
    print("choice:", choice, "reward:", reward) # for debugging

    # remember what action was taken and the corresponding win
    pastRewards[ i ] = ( choice, reward )
    
    # used simply to monitor the run
    runningMean = np.mean( pastRewards[ :i+1, 1 ] ) # i+1 because the upper limit is not included
    plt.scatter( i, runningMean )
    

plt.axhline( y = np.max( arms ) * 10 )
plt.ylim(ymin = 0)
print("best arm: ", best_arm, "(should:", np.argmax ( arms ),"p_win:", np.max ( arms ) ,"),  Avg winning:", runningMean)
print(reward)
print(pastRewards)
plt.show()

elapsed_time = time.time() - starting_time
print("Experiment took {} seconds".format(elapsed_time))

# Making a function out of the code above, in order to launch iter times experiments.
def runExperiment(nb_trials, iter):
    np.random.seed(0)
    arms = np.random.rand( nb_machines )
    np.random.seed()
    for i in range(iter):
        pastRewards = np.zeros( ( nb_trials, 2 ) ) 
        for i in range( nb_trials ):
            best_arm = get_best_arm( pastRewards, range( nb_machines ) )    
            choice = np.random.choice( nb_machines )
            reward = get_reward( arms[ choice ] )
            pastRewards[ i ] = ( choice, reward )
            runningMean = np.mean( pastRewards[ :i+1, 1 ] )
            plt.scatter( i, runningMean )
        plt.axhline( y = np.max( arms ) * 10 )
        plt.ylim(ymin = 0)
        print("best arm: ", best_arm, "(should:", np.argmax ( arms ),"p_win:", np.max ( arms ) ,"),  Avg winning:", runningMean)
        print(reward)
        print(pastRewards)
        plt.show()
        plotRunningMean(i, runningMean, arms)
        plotAvgRatio(i, runningMean / np.max ( arms ), arms)
        plotBoxPlot(pastRewards[:,1])
        # printing the best arm in history
        print("best arm in history:", get_best_arm(pastRewards, range(nb_machines)))

# Plotting the evolution of the average winnings over the trials of an experiment.
def plotRunningMean(i, runningMean, arms):
    plt.scatter( i, runningMean )
    plt.axhline( y = np.max( arms ) * 10 )
    plt.ylim(ymin = 0)
    plt.show()

# Plotting the average ratio average winnings / best possible winnings over 10 or more experiments.
def plotAvgRatio(i, avgRatio, arms):
    plt.scatter( i, avgRatio )
    plt.axhline( y = np.max( arms ) * 10 )
    plt.ylim(ymin = 0)
    plt.show()

# Plotting the winning per machine as a series of box plots or violin plots in a single figure.
def plotBoxPlot(arms):
    plt.boxplot(arms)
    plt.show()

runExperiment(1000, 15)
```

The softmax strategy is a probabilistic strategy that chooses the action with the highest probability. It differs from the epsilon-greedy strategy in the fact that it does not choose the action with the highest average reward, but the action with the highest probability. The advantage of the softmax strategy is that it is more flexible than the epsilon-greedy strategy, since it can be used to explore the environment more or less.
